
Dataviz 01 - ScatterPlot com a relação entre duas variáveis


```python
# Dataviz 01
plt.figure(figsize = (20,8))
sns.relplot(x = 'total_bill', y = 'tip', data = dados)
```

Podemos observar que a medida que aumenta o valor da conta de cada cliente no restaurante, aumenta o valor da gorjeta. Sendo assim pode-se observar que as duas variáveis tem uma correlação positiva.


Dataviz 02 - Scatter Plot com a relação entre duas variáveis e os pontos coloridos pelas categorias de uma terceira variável.


```python
# Dtaviz 02 
plt.figure(figsize = (20,8))
sns.relplot(x = 'total_bill', y = 'tip', hue = 'smoker', data = dados)
```

Temos as mesmas informações trazidas pelo gráfico anterior, porem neste gráfico é possível diferenciar os clientes que são ou não fumantes.


Dataviz 03 - Scatter Plot com a relação entre duas variáveis, os pontos coloridos pelas categorias de uma terceira variável e diferentes símbolos por categoria (gráfico mostrando informação de 3 variáveis diferentes).


```python
# Dataviz 03
plt.figure(figsize = (20,8))
sns.relplot(x = 'total_bill', y = 'tip', hue = 'smoker', style = 'smoker', data = dados)
```

Temos as mesmas informações trazidas pelo gráfico anterior, porem neste gráfico é possível diferenciar os clientes que são ou não fumantes e na categoria de fumantes, podemos ver a diferença entre os símbolos.


Dataviz 04 - Scatter Plot com a relação entre duas variávies, os pontos coloridos pelas categorias de uma variável, diferentes símbolos por categoria e por variável (gráfico mostrando informações de 4 variáveis diferentes).

```python
# Dataviz 04
plt.figure(figsize = (20,8))
sns.relplot(x = 'total_bill', y = 'tip', hue = 'smoker', style = 'time', data = dados)
```

Neste gráfico vemos as mesmas informações anteriore, com uma variável a mais. Agora temos a categoria Smoker em coloração diferente e a time se o cliente foi almoçar ou jantar representados em diferentes símbolos.


Dataviz 05 - Scatter Plot com a relação entre duas variáveis e os pontos coloridos pelas categorias de uma terceira variável, com ajuste na escala de símbolos (gráfico mostrando informações de 3 variáveis diferentes).


```python
# Dataviz 05
plt.figure(figsize = (20,8))
sns.relplot(x = 'total_bill', y = 'tip', hue = 'size', data = dados)
```

Neste gráfico a variável size se torna uma coloração gradiente, representando quantos clientes tem por mesa no restaurante. Sendo a cor mais fraca menos cliente por mesa e a cor mais forte aumenta o número de cliente por mesa.


Dataviz 06 - Scatter Plot com a relação entre duas variáveis e os pontos coloridos pelas categorias de uma terceira variável, com ajuste na escala de símbolos (gráfico mostrando informações de 3 variáveis diferentes).


```python
# Dataviz 06

# Definindo os parâmetros para a paleta Cubehelix
cmap = sns.cubehelix_palette(start=2.8, rot=0.1, hue=0.75, dark=0.2, light=0.8, reverse=True)

plt.figure(figsize = (20,8))
sns.relplot(x = 'total_bill', y = 'tip', hue = 'size', palette=cmap, data = dados)
```

Neste gráfico a escala de cor foi invertida, mais forte menos cliente por mesa e a cor mais fraca aumenta o número de cliente por mesa.


Dataviz 07 - Scatter Plot com a relação entre duas variávies e os pontos com tamanhos definidos pelas categorias de uma terceira variável (gráfico mosntrando informações de 3 variáveis diferentes).

```python
# Dataviz 07
plt.figure(figsize = (20,8))
sns.relplot(x = 'total_bill', y = 'tip', size = 'size', data = dados)
```

Neste gráfico diferentes dos anteriores que foram pela escala de cores, aqui a forma de diferenciar é atráves dos tamanhos das mesas.


Dataviz 08 - Scatter Plot com a relação entre duas variáveis e os pontos com os tamanhos definidos pelas categorias de umaterceira variável, com ajuste da escala dos símbolos (gráfico mostrando informações de 3 variáveis diferentes).


```python
# Dataviz 08
plt.figure(figsize = (20,8))
sns.relplot(x = 'total_bill', y = 'tip', size = 'size', sizes = (15, 200), data = dados)
```

Aqui mostramos o gráfico anterior so que em uma escala maior, ou seja aumentamos o tamanho dos circulos, que se referem as mesas.
